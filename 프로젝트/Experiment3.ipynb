```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data3 = pd.read_csv("Experiment3.csv")
```

```python
data3
```

```python
x = data3["TIME"]
y1 = data3["Target Y"]

plt.plot(x, y1, 'b', label = "Target Y")
plt.xlabel("TIME(s)")
plt.ylabel("Target Y(mm)")
plt.legend(loc="best")
plt.show()
```

```python
y2 = data3["Actual Y"]
plt.plot(x, y2, 'g', label = "Actual Y")
plt.xlabel("TIME(s)")
plt.ylabel("Actual Y(mm)")
plt.legend(loc="best")
plt.show()
```

```python
plt.plot(x, y1, 'b', label = "Target Y")
plt.plot(x, y2, 'g', label = "Actual Y")
plt.xlabel("TIME(s)")
plt.ylabel("Target & Actual Y(mm)")
plt.legend(loc="best")
plt.show()
```

```python
y3 = data3["Following Error"]
plt.plot(x, y3, 'r', label = "Following Error")
plt.xlabel("TIME(s)")
plt.ylabel("Fol err(mm)")
plt.legend(loc="best")
plt.show()
```
